```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml

%matplotlib inline
```

```python
%cd O:\Persons\Kartikeya
df = pd.read_csv('Glioblastoma_Cells_Report_Common_Neurons_P53.csv')
print(df.shape)
df.head(2)
```

```python
df_data = df.drop(['Cell_type'], axis=1)
df.head(2)
```

```python
# Perform PCA and save first 5 PCs
pca = PCA(n_components=5)
pca.fit(df_data)
pca_components = pca.transform(df_data)
df_pca = pd.DataFrame(data=pca_components, columns=[f"PC{i}" for i in range(1, 6)])

df_pca["Cell_type"] = df["Cell_type"]

# Define colors and legend
colors = {"0": "#F1746B", "1": "#36C3D1"}
custom_lines = [Line2D([0], [0], marker='o', label='Scatter', markerfacecolor='#F1746B', markersize=30, color='floralwhite'),
                Line2D([0], [0], marker='o', label='Scatter', markerfacecolor='#36C3D1', markersize=30, color='floralwhite')]
```

```python
# Print summary of PCA results
print("PCA explained variance ratio:", pca.explained_variance_ratio_)
```

```python
# Display PCA dataframe
print(df_pca.head(2))
```

```python

# Plot PCA 1 vs PCA 2
plt.figure(figsize=(30, 25))
plt.scatter(df_pca["PC1"], df_pca["PC2"], c=df_pca["Cell_type"].astype(str).map(colors), s=500)
plt.xlabel('PCA 1', fontsize=60, fontweight='bold', color='black')
plt.ylabel('PCA 2', fontsize=60, fontweight='bold', color='black')
plt.xticks(fontsize=50, fontweight='bold', color='black')
plt.yticks(fontsize=50, fontweight='bold', color='black')
plt.legend(custom_lines, ['Neurons', 'P53'], loc='lower left', shadow=True,
           facecolor='floralwhite', prop={'size': 50})
plt.title(label="PCA: Methylation Neurons vs P53", fontsize=60, fontweight='bold',
          color="black", pad=20)
#plt.savefig("PCA_PCA1vsPCA2_CpG_PatVsHealthy_UPSIT.pdf", format="pdf")
#plt.savefig("PCA_PCA1vsPCA2_CpG_PatVsHealthy_UPSIT.svg", format="svg")
plt.show()
```
