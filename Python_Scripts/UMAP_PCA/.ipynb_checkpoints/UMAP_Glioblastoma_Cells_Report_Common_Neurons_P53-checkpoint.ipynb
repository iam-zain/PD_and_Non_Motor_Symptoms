```python
import umap
import hdbscan
import numpy as np
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.cluster as cluster
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score
%matplotlib inline
```

```python
%cd O:\Persons\Kartikeya
df = pd.read_csv('Glioblastoma_Cells_Report_Common_Neurons_P53.csv')
print(df.shape)
df.head(2)
```

```python
df_targets = df[['Cell_type']]
```

```python
df = df.drop(['Cell_type'], axis=1)
df.head(2)
```

```python
normalized_df=(df-df.min())/(df.max()-df.min())
normalized_df.head()
```

```python
colors = {0:"#F1746B", 1: "#36C3D1"}
```

```python
clusterable_embedding = umap.UMAP(n_neighbors=2, min_dist=0.0, n_components=2,).fit_transform(normalized_df)
```

```python
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], marker='o',label='Scatter',markerfacecolor='#F1746B', markersize=30, color='floralwhite'),
                Line2D([0], [0], marker='o',label='Scatter',markerfacecolor='#36C3D1', markersize=30, color='floralwhite')]
```

```python
plt.scatter(clusterable_embedding[:, 0], clusterable_embedding[:, 1],
c=df_targets['Cell_type'].map(colors), s=500, cmap="Spectral");
plt.rcParams["figure.figsize"] = [30,25]
plt.xlabel('UMAP 1', fontsize=60, fontweight='bold', color='black')
plt.ylabel('UMAP 2', fontsize=60, fontweight='bold', color='black')
plt.xticks(fontsize=50, fontweight='bold', color='black')
plt.yticks(fontsize=50, fontweight='bold', color='black')
plt.legend(custom_lines, ['Neurons', 'P53'], loc='upper left',shadow = True,
           facecolor = 'floralwhite',prop = {'size' : 50})
plt.title(label= "UMAP: Neurons vs P53", fontsize=60, fontweight='bold',
          color='black', pad= 20)
#plt.savefig("UMAP_NonMototSymptoms_PatVsHealthy_Female_Cat.pdf", format="pdf")
#plt.savefig("UMAP_NonMototSymptoms_PatVsHealthy_Female_Cat.svg", format="svg")
plt.show()
```
